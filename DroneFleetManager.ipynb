```python
class drone:
    """A drone with an id, model, battery, status, and missions flown.""";

    def __init__(self, drone_id, model, battery=100, status="Normal", missions_flown=0):
        self.drone_id       = drone_id;
        self.model          = model;
        self.battery        = battery;
        self.status         = status;
        self.missions_flown = missions_flown;

    def assign_mission(self, battery_cost):
        """Assign a mission, and specify the battery cost.""";
        if self.battery < battery_cost:
            return None;
        else:
            self.missions_flown += 1;
            self.battery -= battery_cost;

    def recharge(self, target=100):
        """Recharge the battery to a target level.""";
        self.battery = target;

    def check_battery(self):
        """Prints the battery level.""";
        print(f"{self.battery}%");
        print("WARNING: CRITICAL BATTERY") if self.battery <= 15 else None;

    def status_level(self):
        """Sets the status based on the battery level.""";
        if   self.battery <= 25:
            self.status = "Critical";
            return "Critical, please charge immediately";
        elif self.battery <= 50:
            self.status = "Low";
            return "Low, charging recommended";
        elif self.battery <= 80:
            self.status = "Normal";
            return "Normal, mission-ready";
        else:
            self.status = "Excellent";
            return "Excellent, send on a mission already!";

    def __str__(self):
        """Return a neat string of information.""";
        return f"""{"Drone Report":^50}
{("="*50):^50}

Drone ID        =     D-{self.drone_id}
Model           =     {self.model}
Battery         =     {self.battery}%
Status          =     {self.status}
Missions Flown  =     {self.missions_flown} missions""";


class Fleet:
    """Manages a collection of drone objects.""";

    def __init__(self):
        self.drones = {};

    def add_drone(self, drone_id, drone):
        """Add a drone to the fleet.""";
        self.drones[drone_id] = drone;
        return self;

    def remove_drone(self, drone_id):
        """Remove a drone by ID if it exists."""
        if drone_id in self.drones:
            del self.drones[drone_id]

    def fleet_status(self):
        """Print status of all drones.""";
        for drone_id, drone in self.drones.items():
          print(f"{drone_id}: {drone}");

    def lowest_battery(self):
        """Return the drone with the lowest battery."""
        if not self.drones:
            return None
        return min(self.drones.values(), key=lambda drone: drone.battery).drone_id;

    def most_missions(self):
        """Return the drone with the lowest battery."""
        if not self.drones:
            return None
        return max(self.drones.values(), key=lambda drone: drone.missions_flown).drone_id;

    def __str__(self):
        """Return a neatly formatted fleet report.""";
        lowest_batt_id = self.lowest_battery()
        mvp_id = self.most_missions()

        lowest_battery_drone_info = f"{lowest_batt_id}: {self.drones[lowest_batt_id].battery}%" if lowest_batt_id else "N/A"
        mvp_drone_info = f"{mvp_id}" if mvp_id else "N/A"

        return f"""{"Fleet Report":^50}
{("="*50):^50}

  Number of Drones  =     {len(self.drones)}
  Lowest Battery    =     {lowest_battery_drone_info}
  MVP               =     {mvp_drone_info}: {self.drones[mvp_id].missions_flown} missions
  Total Missions    =     {sum(drone.missions_flown for drone in self.drones.values())} missions""";
```
